```python
import matplotlib.pyplot as plt
from matplotlib import ticker
import numpy as np
import pandas as pd
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import keras
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation, Flatten
from keras.layers.convolutional import Convolution2D, MaxPooling2D
from keras.optimizers import SGD, RMSprop, Adam
```

```python

data["Classe_bonus_malus"] = data["Classe_bonus_malus"].astype('category')
# on définit x et y
y = data["Classe_bonus_malus"].cat.codes
x = data.drop(["Classe_bonus_malus"], axis=1)
data.describe()
```

```python
#Resampling
import scipy.stats as stats
X0=x.values
Y0=y.values
Xb=np.concatenate((X0,X0),axis=0)
Yb=np.concatenate((Y0,Y0),axis=0)

X11=Xb[:,0:3]+stats.norm.rvs(loc = 0.5, scale = 1, size = (373*2,3))
X22=Xb[:,3:6]+stats.uniform.rvs(size=(373*2,3),loc = 0, scale = 0.5)
X1=np.concatenate((X11,X22),axis=1)
X=np.concatenate((Xb,X1),axis=0)
Y=np.concatenate((Yb,Yb),axis=0)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=1/3)

# on ne prend que les colonnes quantitatives
#x = data.select_dtypes(np.number).drop(["X1","X2","X3","X4","X5","X6","X7"],axis=1)
#np.shape(X_train)
#y_train
```

```python
from sklearn.preprocessing import LabelBinarizer
y_dense_train = LabelBinarizer().fit_transform(y_train)
y_dense_test = LabelBinarizer().fit_transform(y_test)
np.shape(y_dense_test)
np.shape(y_dense_test)
y_dense_test
```

```python
modele = keras.Sequential([
layers.Dense(20, activation = 'linear', input_shape = [X_train.shape[1]]),
layers.Dropout(0.1, seed = 3),
layers.Dense(40, activation = 'swish'),
#layers.Dense(4, activation = 'relu'),
#layers.Dense(64, activation = 'swish'),
#layers.Dense(64, activation = 'relu'),
#layers.Dense(20, activation = 'linear'),
layers.Dense(4, activation='sigmoid')])
```

```python
modele.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
#optimiseur = tf.keras.optimizers.RMSprop(learning_rate = 0.1)
#modele.compile(loss = tf.keras.losses.MeanSquaredError(),
#optimizer = optimiseur,
#metrics = ['mae']) 
```

```python
modele.summary()
```

```python
history = modele.fit (X_train, y_dense_train, epochs = 100, batch_size=128, validation_split = 0.2, verbose=1)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','validation'], loc='upper left')
plt.show()
```

```python
pred=np.around(modele.predict(X_test),0)
y_pred=pred.astype(int)
np.shape(y_pred)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import multilabel_confusion_matrix

matrice_confusion2 = multilabel_confusion_matrix(y_dense_test, y_pred)
matrice_confusion2
#print(76/249)
```

```python
labelpredict=modele.predict(X_test)
labelpredict=labelpredict.round()
score = modele.evaluate( X_test, y_dense_test, verbose=0)
score
```

```python
from sklearn import svm, datasets
import sklearn.model_selection as model_selection
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
rbf = svm.SVC(kernel='rbf', gamma=0.5, C=0.1).fit(X_train, y_train)
poly = svm.SVC(kernel='poly', degree=3, C=1).fit(X_train, y_train)
```

```python
poly_pred = poly.predict(X_test)
rbf_pred = rbf.predict(X_test)
```

```python
poly_accuracy = accuracy_score(y_test, poly_pred)
poly_f1 = f1_score(y_test, poly_pred, average='weighted')
print('Accuracy (Polynomial Kernel): ', "%.2f" % (poly_accuracy*100))
print('F1 (Polynomial Kernel): ', "%.2f" % (poly_f1*100))
```

```python
rbf_accuracy = accuracy_score(y_test, rbf_pred)
rbf_f1 = f1_score(y_test, rbf_pred, average='weighted')
print('Accuracy (RBF Kernel): ', "%.2f" % (rbf_accuracy*100))
print('F1 (RBF Kernel): ', "%.2f" % (rbf_f1*100))
```

```python
matrice_confusion2_rbf = confusion_matrix(y_test, rbf_pred)
matrice_confusion2_poly = confusion_matrix(y_test, poly_pred)
print(matrice_confusion2_rbf, '=matrice_confusion2_rbf')
print(matrice_confusion2_poly, '=matrice_confusion2_poly')
```
